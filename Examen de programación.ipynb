# Pregunta 1

```python
def invertir_numero(n):
    numero = 0
    while n != 0:
        numero = 10*numero+n % 10
        n //= 10
    return numero


numero = int(input('Ingresar numero a invertir : '))
numero_invertido = invertir_numero(numero)
print(f"El número {numero} es {numero_invertido} al invertirse")
```

# Pregunta 2

```python
def promedio_palabras(oracion):
    filtrar_puntos = ''.join(filter(lambda x: x not in '".,;!-', oracion))
    palabras = [palabra for palabra in filtrar_puntos.split() if palabra]
    promedio = sum(map(len, palabras)) / len(palabras)
    return print(f'El promedio de las palabras de la oración es {promedio}')


oracion = input("Escriba una oración: ")
promedio_palabras(oracion)
```

```python
def Sumar_num_cadena(numero1, numero2):
    n1=0
    n2=0
    for i in numero1:
        n1=n1*10+float(i)
    for j in numero2:
        n2=n2*10+float(j)
        
    return print(f'La suma de ambos numeros es : {n1 + n2}')


numero1 = input('Ingrese el primer numero :')
numero2 = input('Ingrese el segundo numero :')
Sumar_num_cadena(numero1,numero2)
```

# PREGUNTA 5

```python
def Palindromo(cadena: str, low: int, high: int) -> bool:
    while low < high:
        if cadena[low] != cadena[high]:
            return False
        low += 1
        high -= 1
    return True

def Remover_caracter_palindromo(cadena: str) -> int:

    low = 0
    high = len(cadena) - 1
 
    while low < high:
 
        if cadena[low] == cadena[high]:
            low += 1
            high -= 1
        else:
 
            if Palindromo(cadena, low + 1, high):
                return low
 
            if Palindromo(cadena, low, high - 1):
                return high
            return -1
 

cadena = input('Introducir texto : ')
verificar_cadena = Remover_caracter_palindromo(cadena)   
if verificar_cadena == -1:
    print("No es posible")
elif verificar_cadena == -2:
    print("Posible sin eliminar ningún carácter")
else:
    print("Posible eliminando el carácter en el índice", verificar_cadena)
```

```python
def Verificacion_matriz(Matriz):

    X = (all(Matriz[i] <= Matriz[i + 1] for i in range(len(Matriz) - 1)) or
            all(Matriz[i] >= Matriz[i + 1] for i in range(len(Matriz) - 1)))
    
    return print("La matriz es monótona" if X == True else "La matriz no es monótona")
  
Elementos_matriz = int(input("Introducir numero de elementos de matriz : "))
Matriz=[]
for x in range(Elementos_matriz):
    valor=int(input("Ingrese un valor entero:"))
    Matriz.append(valor)

Verificacion_matriz(Matriz)
```

# PREGUNTA 7

```python
def reordenar(Matriz):
    
    
    print(f'Matriz original : {Matriz}')
    indice = 0
    for i in Matriz:
        if i:
            Matriz[indice] = i
            indice = indice + 1
            
    for i in range(indice, len(Matriz)):
        Matriz[i] = 0
    
    return print (f'Matriz con 0 al final : {Matriz}')


Elementos_matriz = int(input("Introducir numero de elementos de matriz : "))
Matriz=[]
for x in range(Elementos_matriz):
    valor=int(input("Ingrese un valor entero:"))
    Matriz.append(valor)

reordenar(Matriz)
```

```python
def Llenar_segundo(Matriz):
    print(f'Matriz inicial : {Matriz}')
    primero = True
    for i in range(len(Matriz)):
        if primero:
            if Matriz[i] != None:
                Matriz[:i] = [Matriz[i] for _ in range(i)]
                primero = False
        if Matriz[i] == None:
            Matriz[i] = Matriz[i-1]
    return print(f'Matriz sin None : {Matriz}')

Matriz = [2,None,3,None,4,None,None]
Llenar_segundo(Matriz)
```

# PREGUNTA 9

```python
def matriz_union_diferente(oracion1, oracion2):
    i = {}
    
    sep_oracion1 = set(oracion1.split())
    sep_oracion2 = set(oracion2.split())
    matriz_igual = sep_oracion2 & sep_oracion1
    
    for palabra in oracion1.split():
        i[palabra] = i.get(palabra, 0) + 1
    
    for palabra in oracion2.split():
        i[palabra] = i.get(palabra, 0) + 1
  
    matriz_diferente = [palabra for palabra in i if i[palabra] == 1]

    return print(f'Matriz que tiene las palabras que aparecen en una oración y no en la otra : {matriz_diferente} \nMatriz con las palabras en común : {matriz_igual} ')
  

oracion1 = "Feliz mes"
oracion2 = "Feliz mes patrio Perú"
  
matriz_union_diferente(oracion1, oracion2)
```

# PREGUNTA 10

```python
def es_primo(n):
    if n <= 1 :
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True


def Imprimir_numeros_primos(n):
    for i in range(2, n + 1):
        if es_primo(i):
            print(i, end = " ")

numero = int(input('Escribir número : '))

Imprimir_numeros_primos(numero)
```
