# Pregunta 1

```python
def invertir_numero(n):
    numero = 0
    while n != 0:
        numero = 10*numero+n % 10
        n //= 10
    return numero


numero = int(input('Ingresar numero a invertir : '))
numero_invertido = invertir_numero(numero)
print(f"El número {numero} es {numero_invertido} al invertirse")
```

# Pregunta 2

```python
def promedio_palabras(oracion):
    filtrar_puntos = ''.join(filter(lambda x: x not in '".,;!-', oracion))
    palabras = [palabra for palabra in filtrar_puntos.split() if palabra]
    promedio = sum(map(len, palabras)) / len(palabras)
    return print(f'El promedio de las palabras de la oración es {promedio}')


oracion = input("Escriba una oración: ")
promedio_palabras(oracion)
```

# Pregunta 3

```python
def sumar_dos_numeros(num1,num2):
    sum = float(num1) + float(num2)
    return f'La suma de dos numeros es {sum}'

num1 = input('Escribir primer número')
num2 = input('Escribir segundo número')

sumar_dos_numeros(num1,num2)
```
